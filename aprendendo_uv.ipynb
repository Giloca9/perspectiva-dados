# Comandos Git

1. `git clone`: baixa o repositório do GitHub na sua máquina
    (Exemplo: git clone https://github.com/Giloca9/perspectiva-dados.git )

2. `git status`: Mostra o estado dos arquivos

3. `git diff`: Mostra o que foi alterado

4. `git add`: Escolhe os arquivos que quero salvar

5. `git commit -m " "`: Salva o que foi escolhido no add

6. `git push`: Envia tudo que foi commitado para o GitHub

# Atalho Teclado

1. Ctrl + Alt + T
Abre o terminal no Linux.

2. Ctrl + Shift + V
Cola texto no terminal.

3. Ctrl + H
Normalmente mostra ou oculta arquivos ocultos, dependendo do programa.

# Comandos Terminal

1. `ls`: lista todos os arquivos na pasta
2. `cd`: Troca de diretório
3. `cat`: Abre um arquivo direto do terminal
