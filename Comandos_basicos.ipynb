# Comandos Git

1. `git clone`: baixa o repositório do GitHub na sua máquina  
   > **Exemplo:** `git clone https://github.com/Giloca9/perspectiva-dados.git`

2. `git status`: mostra a situação atual dos arquivos (arquivos com alterações, arquivos novos, quais estão prontos para commit, etc.)

3. `git diff`: mostra exatamente o que foi alterado

4. `git add`: separa os arquivos para o commit  
   > **Exemplo:** `git add nome_do_arquivo`  
   > **Exemplo:** `git add .` (separa todos os arquivos da pasta)

5. `git commit -m " "`: salva localmente tudo o que foi separado no `add`  
   > **Exemplo:** `git commit -m "Deletei arquivos desnecessários"`

6. `git push`: envia tudo que foi commitado para o GitHub

7. `git restore`: desfaz as alterações que você fez no arquivo  
   > **Exemplo:** `git restore nome_do_arquivo`

8. `git restore --staged`: tira um arquivo da área de preparação (`add`)  
   > **Exemplo:** `git restore --staged nome_do_arquivo`

9. `git log`: mostra o histórico de commits

## Arquivo importante do Git

1. `.gitignore`: arquivo para dizer quais arquivos ou pastas devem ser ignorados

# Atalho Teclado

1. Ctrl + Alt + T
Abre o terminal no Linux.

2. Ctrl + Shift + V
Cola texto no terminal.

3. Ctrl + H
Normalmente mostra ou oculta arquivos ocultos, dependendo do programa.

# Comandos Terminal

1. `ls`: lista todos os arquivos e pastas do diretório atual

2. `ls -a`: lista todos os arquivos da pasta, inclusive os arquivos ocultos

3. `cd`: muda de diretório dentro do terminal (modo hacker sem mouse)  
   > **Exemplo:** `cd nome_da_pasta`

4. `cat`: mostra o conteúdo de um arquivo no terminal  
   > **Exemplo:** `cat README.md`

5. `touch`: cria um arquivo vazio  
   > **Exemplo:** `touch nome_do_arquivo`
   
6. `rm`: remove arquivo  
   > **Exemplo:** `rm nome_do_arquivo`

# O que é `.bashrc`

1. `.bashrc`: é um arquivo de configuração do terminal

2. Serve para armazenar personalizações do terminal, como alias (comandos personalizados) e configurações que são carregadas automaticamente quando o terminal é aberto.

# Comandos uv

`uv`: ferramenta que gerencia o projeto Python

1. `uv add`: adiciona pacote ao projeto
   > **Exemplo:** `uv add pandas`

2. `uv run`: roda comando ou script 
   > **Exemplo:** `uv run jupyter notebook`


# Editor de texto vi

## Modos do vi

**Modo de navegação**: andar pelo arquivo e executar comandos\
**Modo de edição**:  escrever e alterar o texto

## Comandos vi

1. `vi`: abre o editor de texto no terminal  
> **Exemplo**: `vi nome_do_arquivo.txt`

2. `i`: entra no modo de edição (insert), para começar a digitar

3. `o`: cria uma nova linha abaixo da linha atual e entra no modo de edição

4. `Esc`: sai do modo de edição e volta para o modo de navegação

5. `:w`: salva as alterações do arquivo

6. `:q`: sai do editor

7. `:q!`: sai do editor sem salvar as alterações feitas
