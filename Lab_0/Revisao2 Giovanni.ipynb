```python
np.random.seed(16)
matriz1 = np.random.normal(0, 1, size=(5, 5))
print(matriz1)
print()

# Soma linhas

print(np.sum(matriz1[0,:]))
print(np.sum(matriz1[1,:]))
print(np.sum(matriz1[2,:]))
print(np.sum(matriz1[3,:]))
print(np.sum(matriz1[4,:]))
print()

# soma colunas

print(np.sum(matriz1[:,0]))
print(np.sum(matriz1[:,1]))
print(np.sum(matriz1[:,2]))
print(np.sum(matriz1[:,3]))
print(np.sum(matriz1[:,4]))
print()

#DIagonal principal

diagonal_principal = np.array([matriz1[0,0],matriz1[1,1], matriz1[2,2], matriz1[3,3], matriz1[4,4]] )
print(diagonal_principal)
```

```python
def ex3 (vetorA, vetorB):
    produto_escalar = np.dot(vetorA,vetorB)
    distancia_euclidiana = np.linalg.norm(vetorA - vetorB)
    return produto_escalar, distancia_euclidiana

vetora = np.array([1,2,3])
vetorb = np.array([4,5,6])

print(ex3(vetora,vetorb))
```

```python
np.random.seed(16)
matriz = np.random.randint(-5, 6, size=(4, 4))
print(matriz)
print()

print(np.linalg.det(matriz))
print()
# Det diferente de 0, logo é inversível

inversa = np.linalg.inv(matriz)
print(inversa)
print()

print(np.round(matriz@inversa,3))
#virou identidade
```

```python
np.random.seed(16)
M = np.random.randint(0, 10, size=(5, 5))
print(M)
print()

B = 0.5*(M + M.T)
print(B)
print()

print(np.linalg.eig(B))
print()

autovalores = np.linalg.eig(B)[0]
print(autovalores)
print()

autovetores = np.linalg.eig(B)[1]
print(autovetores)
print()

for i in range(len(autovalores)):
    v = autovetores[:, i]
    lamb = autovalores[i]

    print(np.round(B @ v,2))
    print(np.round(lamb * v,2))
    print()
```

```python
np.random.seed(16)
A = np.random.randint(0, 10, size=(3, 3))
y = np.random.randint(0, 10, size=(3))

print(A)
print()

print(y)
print()

x = np.linalg.solve(A,y)
print(x)
print()

np.round(np.linalg.norm(A @ x - y),4)
```

```python
np.random.seed(16)

dados = np.random.normal(0, 1, size=(100, 2))
S = np.cov(dados, rowvar=False)

print(S)
print()

autovalores = np.linalg.eig(S)[0]
print(autovalores)
print()

autovetores = np.linalg.eig(S)[1]
print(autovetores)

```

```python
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("iris.csv")

plt.hist(df["sepal.length"])
plt.title("Histograma sepal.length")
plt.xlabel("sepal.length")
plt.ylabel("Frequência")


```

```python

dados_setosa = df.query("variety == 'Setosa'")
plt.scatter(dados_setosa["sepal.length"], dados_setosa["sepal.width"], label="Setosa")

dados_versicolor = df.query("variety == 'Versicolor'")
plt.scatter(dados_versicolor["sepal.length"], dados_versicolor["sepal.width"], label="Versicolor")

dados_virginica = df.query("variety == 'Virginica'")
plt.scatter(dados_virginica["sepal.length"], dados_virginica["sepal.width"], label="Virginica")

plt.legend()
```

```python
plt.scatter(df["petal.length"], df["petal.width"], s=df["sepal.length"])

plt.xlabel("petal.length")
plt.ylabel("petal.width")
plt.title("Gráfico de dispersão")
```

```python

plt.subplot(1, 3, 1)
plt.hist(df.query("variety == 'Setosa'")["petal.length"])
plt.title("Setosa")

plt.subplot(1, 3, 2)
plt.hist(df.query("variety == 'Versicolor'")["petal.length"])
plt.title("Versicolor")

plt.subplot(1, 3, 3)
plt.hist(df.query("variety == 'Virginica'")["petal.length"])
plt.title("Virginica")

```

```python
plt.boxplot([df["sepal.length"], df["sepal.width"], df["petal.length"], df["petal.width"]],
    labels=["sepal.length", "sepal.width", "petal.length", "petal.width"])

# petal.length tem a maior variabilidade, pois apresentou maior amplitude e maior dispersão no boxplot.”
```

```python
medias = df.groupby("variety")[["sepal.length", "sepal.width", "petal.length", "petal.width"]].mean()
print(medias)
print()

plt.figure(figsize=(15,4))

plt.subplot(1, 3, 1)
plt.bar(["sepal.length", "sepal.width", "petal.length", "petal.width"],
        medias.loc["Setosa"])
plt.title("Setosa")
plt.ylabel("Média")

plt.subplot(1, 3, 2)
plt.bar(["sepal.length", "sepal.width", "petal.length", "petal.width"],
        medias.loc["Versicolor"])
plt.title("Versicolor")
plt.ylabel("Média")

plt.subplot(1, 3, 3)
plt.bar(["sepal.length", "sepal.width", "petal.length", "petal.width"],
        medias.loc["Virginica"])
plt.title("Virginica")
plt.ylabel("Média")

```
