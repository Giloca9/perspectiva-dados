```python
distribuicoes = {
    "Normal": {"params": {"loc": 0, "scale": 1}, "dados": None, "estatisticas": None},
    "Exponencial": {"params": {"scale": 2}, "dados": None, "estatisticas": None},
    "Uniforme": {"params": {"low": -1, "high": 1}, "dados": None, "estatisticas": None}
}
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
np.random.seed(16)

amostra_normal = np.random.normal(0,1,size=1000)
amostra_unifome = np.random.uniform(-1,1,size= 1000)
amostra_exp = np.random.exponential(2, size = 1000)

distribuicoes["Normal"]["dados"] = amostra_normal
distribuicoes["Exponencial"]["dados"] = amostra_exp
distribuicoes["Uniforme"]["dados"] = amostra_unifome
```

```python
est_normal = {
    "Média": None,
    "Mediana": None,
    "Variancia": None,
    "DP": None
}

est_normal["Média"] = np.mean(amostra_normal)
est_normal["Mediana"] = np.median(amostra_normal)
est_normal["Variancia"] = np.var(amostra_normal)
est_normal["DP"] = np.sqrt(est_normal["Variancia"])

distribuicoes["Normal"]["estatisticas"] = est_normal
```

```python
distribuicoes["Normal"]["estatisticas"]
```

```python
est_exp = {
    "Média": None,
    "Mediana": None,
    "Variancia": None,
    "DP": None
}

est_exp["Média"] = np.mean(amostra_exp)
est_exp["Mediana"] = np.median(amostra_exp)
est_exp["Variancia"] = np.var(amostra_exp)
est_exp["DP"] = np.sqrt(est_exp["Variancia"])

distribuicoes["Exponencial"]["estatisticas"] = est_exp
```

```python
distribuicoes["Exponencial"]["estatisticas"]
```

```python
est_uni = {
    "Média": None,
    "Mediana": None,
    "Variancia": None,
    "DP": None
}

est_uni["Média"] = np.mean(amostra_unifome)
est_uni["Mediana"] = np.median(amostra_unifome)
est_uni["Variancia"] = np.var(amostra_unifome)
est_uni["DP"] = np.sqrt(est_uni["Variancia"])

distribuicoes["Uniforme"]["estatisticas"] = est_uni
```

```python
distribuicoes["Uniforme"]["estatisticas"]
```

```python
poisson  = {
    "params": {"scale": 2}, "dados": None, "estatisticas": None}

distribuicoes["Poisson"] = poisson

print(distribuicoes["Poisson"])
```

```python
np.random.seed(16)

amostra_poisson = np.random.poisson(2,size=1000)

distribuicoes["Poisson"]["dados"] = amostra_poisson
```

```python
est_poi = {
    "Média": None,
    "Mediana": None,
    "Variancia": None,
    "DP": None
}

est_poi["Média"] = np.mean(amostra_poisson)
est_poi["Mediana"] = np.median(amostra_poisson)
est_poi["Variancia"] = np.var(amostra_poisson)
est_poi["DP"] = np.sqrt(est_poi["Variancia"])

distribuicoes["Poisson"]["estatisticas"] = est_poi

print(distribuicoes["Poisson"]["estatisticas"])
```

```python
df = pd.DataFrame({distribuicao: distribuicoes[distribuicao]["estatisticas"] for distribuicao in distribuicoes.keys()}).T
print(df)
```

```python
distribuicoes["Normal"]["estatisticas"]["minimo"] = np.min(amostra_normal)
distribuicoes["Normal"]["estatisticas"]["q1"] = np.percentile(amostra_normal, 25)
distribuicoes["Normal"]["estatisticas"]["q3"] = np.percentile(amostra_normal, 75)
distribuicoes["Normal"]["estatisticas"]["maximo"] = np.max(amostra_normal)

distribuicoes["Exponencial"]["estatisticas"]["minimo"] = np.min(amostra_exp)
distribuicoes["Exponencial"]["estatisticas"]["q1"] = np.percentile(amostra_exp, 25)
distribuicoes["Exponencial"]["estatisticas"]["q3"] = np.percentile(amostra_exp, 75)
distribuicoes["Exponencial"]["estatisticas"]["maximo"] = np.max(amostra_exp)

distribuicoes["Uniforme"]["estatisticas"]["minimo"] = np.min(amostra_unifome)
distribuicoes["Uniforme"]["estatisticas"]["q1"] = np.percentile(amostra_unifome, 25)
distribuicoes["Uniforme"]["estatisticas"]["q3"] = np.percentile(amostra_unifome, 75)
distribuicoes["Uniforme"]["estatisticas"]["maximo"] = np.max(amostra_unifome)

distribuicoes["Poisson"]["estatisticas"]["minimo"] = np.min(amostra_poisson)
distribuicoes["Poisson"]["estatisticas"]["q1"] = np.percentile(amostra_poisson, 25)
distribuicoes["Poisson"]["estatisticas"]["q3"] = np.percentile(amostra_poisson, 75)
distribuicoes["Poisson"]["estatisticas"]["maximo"] = np.max(amostra_poisson)
```

```python
df = pd.DataFrame({distribuicao: distribuicoes[distribuicao]["estatisticas"] for distribuicao in distribuicoes.keys()}).T

print(df)
```

```python
plt.boxplot([amostra_normal, amostra_exp, amostra_unifome, amostra_poisson],
            labels=["Normal", "Exponencial", "Uniforme", "Poisson"])

plt.title("Boxplot das distribuições simuladas")
plt.xlabel("Distribuições")
plt.ylabel("Valor")
```

```python
teorico_normal = distribuicoes['Normal']["params"]["loc"]

teorico_exp = distribuicoes["Exponencial"]["params"]["scale"]

teorico_unif = (distribuicoes["Uniforme"]["params"]["low"] + distribuicoes["Uniforme"]["params"]["high"])/2

teorico_poi = distribuicoes["Poisson"]["params"]["scale"]

val_teorico = np.array([teorico_normal, teorico_exp, teorico_unif, teorico_poi])
print(val_teorico)

amostral_normal = distribuicoes["Normal"]["estatisticas"]["Média"]

amostral_exp = distribuicoes["Exponencial"]["estatisticas"]["Média"]

amostral_unif = distribuicoes["Uniforme"]["estatisticas"]["Média"]

amostral_poi = distribuicoes["Poisson"]["estatisticas"]["Média"]

val_amostral = np.array([amostral_normal, amostral_exp, amostral_unif, amostral_poi])
print(val_amostral)

dif = val_teorico - val_amostral
print(dif)

```

```python
df_10 = pd.DataFrame({
    "Valor Teorico": val_teorico,
    "Valor Amostral": val_amostral},
    index=["Normal", "Exponencial", "Uniforme", "Poisson"])

print(df_10)

df_10.plot(kind="bar")
plt.title("Comparação entre média teórica e média amostral")
plt.xlabel("Distribuições")
plt.ylabel("Valor")
```
